# 4. Data Cleaning

# Price & Discount Columns Need Numeric Conversion

```python
# Price & Discount Columns Need Numeric Conversion

data['discounted_price'] = data['discounted_price'].astype(str).str.replace('₹','').str.replace(',','')

data['actual_price'] = data['actual_price'].astype(str).str.replace('₹','').str.replace(',','')

data['discount_percentage'] = data['discount_percentage'].astype(str).str.replace('%','')

# Converting to numeric float format

data['discounted_price'] = data['discounted_price'].astype(float)
data['actual_price'] = data['actual_price'].astype(float)
data['discount_percentage'] = data['discount_percentage'].astype(float)
data.head()
```

# Rating & Rating Count Require Fixing

```python
# Removing (,) from rating_count
data['rating_count'] = data['rating_count'].astype(str).str.replace(',','')

# Convert to numeric, handles 'nan' strings properly
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')

# Fill NaN with 0 (no reviews = 0 count)
data['rating_count'] = data['rating_count'].fillna(0)

# Convert to integer
data['rating_count'] = data['rating_count'].astype(int)

data['rating'] = pd.to_numeric(data['rating'], errors='coerce')

# Fill NaN with median (0 is not a valid rating on a 1–5 scale)
data['rating'] = data['rating'].fillna(data['rating'].median())

data['rating'] = data['rating'].astype(float)

data.info()

data.isnull().sum()
```

# Category Field Is Long & Fragmented


```python
# Extract only the first category

data['category'] = data['category'].astype(str).str.split('|').str[0]

#Remove extra spaces if present

data['category'] = data['category'].str.strip()

data.head()
```

```python
# Removing extra spaces from retained columns only
data['product_name'] = data['product_name'].str.strip()
```

```python
# Dropping columns which are not needed

data = data.drop(['img_link','product_link','review_content','review_title',
                  'user_id','user_name','about_product'], axis=1)

data.head()
```

# 5. Feature Engineering

```python
# Extracting the discount amount

data['discount_amount'] = data['actual_price'] - data['discounted_price']

# Extracting high discount flag: 1 = high discount (>=40%), 0 = low
data['high_discount'] = data['discount_percentage'].apply(lambda x: 1 if x >= 40 else 0)

data.head()
```

### Creating Discount-Based Features

To better understand how pricing strategies influence sales behavior, we created two additional variables.

First, we calculated the **discount amount**, which represents the actual price reduction offered to customers. This helps measure how aggressive the pricing strategy is for each product.

Next, we introduced a **high_discount flag**. Products with a discount of 40% or more were marked as high-discount items. This allows us to compare whether heavily discounted products behave differently from moderately discounted ones.

These features make it easier to analyze the relationship between discounting and customer purchasing patterns.

```python
# Extacting price range Bucketed ranges (0–500, 500–1000, etc.)
bins = [0,500,1000,1500,2000,3000,5000,10000,float('inf')]

labels = [
    '0-500',
    '500-1000',
    '1000-1500',
    '1500-2000',
    '2000-3000',
    '3000-5000',
    '5000-10000',
    '10000+'
]


data['price_range'] = pd.cut(data['discounted_price'], bins=bins , labels=labels , right=False)

data.head()
```

### Price Range Segmentation

To make pricing patterns easier to analyze, products were grouped into defined price ranges (e.g., 0–500, 500–1000, etc.).

This allows us to quickly understand how products are distributed across different pricing tiers and helps identify whether customer activity is concentrated in budget, mid-range, or premium segments. Such segmentation supports more informed pricing and marketing decisions.

```python
# Rating buckets: group ratings into 4 categories (<2, 2–3, 3–4, 4–5)
bins = [0, 2, 3, 4, 5]
labels = ['<2', '2-3', '3-4', '4-5']

data['rating_buckets'] = pd.cut(data['rating'], bins=bins, labels=labels, right=False)

# Extracting net saving percent
data['net_saving_percent'] = round((1 - data['discounted_price'] / data['actual_price']) * 100, 2)
data.head()
```

### Rating Segmentation and Savings Analysis

Customer ratings were grouped into categories to better evaluate product performance levels, from low-rated to highly rated items. This helps identify how product quality perception varies across the dataset.

In addition, a net saving percentage was calculated to measure the effective price customers pay compared to the original price. This provides a clearer view of the actual value offered through discounts and supports analysis of how savings may influence customer satisfaction.

# DESIGN SQL TABLES (Schema + ERD)



After completing the analysis, the next step is to structure this data in a relational format so it can be stored, queried, and maintained efficiently in a database environment.

This involves defining tables, relationships, and keys that represent products, categories, pricing, and customer engagement metrics.

### Objective of the Data Model

The goal of the schema is to transform the analytical dataset into a structured format suitable for real-world use cases such as reporting, dashboards, and transactional systems.

The design focuses on:
- Avoiding data redundancy
- Maintaining clear relationships between entities
- Supporting scalable querying and analysis

```python
import sqlite3 as sp

connt = sp.connect('amazon.db')
cursor = connt.cursor()
```

```python
cursor.executescript("""CREATE TABLE IF NOT EXISTS products(
product_id TEXT PRIMARY KEY,
product_name TEXT,
category TEXT,
discounted_price REAL,
actual_price REAL,
discount_percentage REAL
);

CREATE TABLE IF NOT EXISTS reviews(
review_id TEXT PRIMARY KEY,
product_id TEXT,
rating REAL,
rating_count INTEGER,
FOREIGN KEY (product_id) REFERENCES products(product_id)
);
""")

connt.commit()
connt.close()
```

# TOP CATEGORIES BY REVIEW VOLUME

```python
# TOP CATEGORIES BY REVIEW VOLUME

query = """
SELECT
p.category,
SUM(r.rating_count) AS total_reviews
FROM products p
JOIN reviews r
ON p.product_id = r.product_id
GROUP BY p.category
ORDER BY total_reviews DESC;
"""

pd.read_sql(query,connt)
```

# Best-rated categories (with minimum reviews)

```python
# Best-rated categories (with minimum reviews)

query1 = """
SELECT p.category,
ROUND(AVG(r.rating),2) AS avg_rating,
SUM(r.rating_count) AS total_reviews

FROM products p
JOIN reviews r
ON p.product_id = r.product_id
WHERE r.rating IS NOT NULL
GROUP by p.category
HAVING SUM(r.rating_count) >= 100
ORDER by avg_rating DESC;
"""

pd.read_sql(query1,connt)
```
